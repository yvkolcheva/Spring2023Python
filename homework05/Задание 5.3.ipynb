```python
#Задание 1
import numpy as np
import numpy.linalg
import matplotlib.pyplot as plt

def f(x):
    return np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-1/2 * x)

def appr(n):
    dot = np.linspace(1, 15, n)
    a = np.array([[x**j for j in range(n)] for x in dot])        
    b = f(dot)
    w = np.linalg.solve(a, b)  
    return np.array([sum([w[j] * x**j for j in range(n)]) for x in dot])


x = np.linspace(1, 15, 150)
plt.grid()
plt.title('Функция')
plt.plot(x, f(x))
plt.xlabel('Ось Оx')
plt.ylabel('Ось Оy')
plt.show()

n = 15
plt.grid()
plt.title('Точки, по которым стpоился, n=15')
plt.plot(np.linspace(1, 15, n), f(np.linspace(1, 15, n)), '*')
plt.xlabel('Ось Оx')
plt.ylabel('Ось Оy')
plt.show()

plt.grid()
plt.title('Полиномиальная аппроксимация, n=15')
plt.plot(np.linspace(1, 15, n), appr(n))
plt.xlabel('Ось Оx')
plt.ylabel('Ось Оy')
plt.show()
```

```python
#Задание 2
import numpy as np
import numpy.linalg
import matplotlib.pyplot as plt

def f(x):
    return np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-1/2 * x)

def appr(n):
    dot = np.linspace(1, 15, n)
    a = np.array([[x**j for j in range(n)] for x in dot])        
    b = f(dot)+ np.random.normal(0, 0.01)
    w = np.linalg.solve(a, b)  
    return np.array([sum([w[j] * x**j for j in range(n)]) for x in dot])


x = np.linspace(1, 15, 150)
plt.grid()
plt.title('Функция')
plt.plot(x, f(x))
plt.xlabel('Ось Оx')
plt.ylabel('Ось Оy')
plt.show()

n = 15
plt.grid()
plt.title('Точки, по которым стpоился, n=15')
plt.plot(np.linspace(1, 15, n), f(np.linspace(1, 15, n)), '*')
plt.xlabel('Ось Оx')
plt.ylabel('Ось Оy')
plt.show()

plt.grid()
plt.title('Полиномиальная аппроксимация, n=15')
plt.plot(np.linspace(1, 15, n), appr(n))
plt.xlabel('Ось Оx')
plt.ylabel('Ось Оy')
plt.show()
```
