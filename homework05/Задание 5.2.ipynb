```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
```

```python
#Часть 1
x = np.linspace(-1, 1, 250)
y = (x*x)*np.sin(1/(x*x)) + x
plt.xlabel(u'Ось x', family='verdana')
plt.ylabel(u'Ось y', family='verdana')
plt.title(u'Первая функция', family='verdana')
plt.plot(x, y)
```

```python
a = 1/(1+x*x)
b= 1/(x*x)
x = np.linspace(5, 25, 100)
y = np.abs(1 - b/a)


plt.xlabel(u'Ось x', family='verdana')
plt.ylabel(u'Ось y', family='verdana')
plt.title(u'Вторая функция', family='verdana')

plt.semilogy(x, y)
plt.grid()
```

```python
from scipy.special import erfi, gammasgn

plt.title(u'Третья функция', family='verdana')
x1 = np.linspace(-2, 2, 100)
y1 = erfi(x1)
x2 = x1
y2 = gammasgn(x2)

plt.grid()
plt.plot(x1, y1, ':b', label='erfi')
plt.plot(x2, y2, '-g', label='gammasgn')
plt.legend()
```

```python
#Часть 2
x = np.linspace(-6*np.pi, 6*np.pi, 100)
y = np.sin(x) / x



fig = plt.figure(figsize=(15, 5))
ax = fig.add_subplot()
ax.plot(x, y, color='g', linestyle='-', marker='$\u03C0$', markevery=10)

ax.set_xlabel('Ось x')
ax.set_ylabel('Ось у')
ax.set_title('Функция')

ax.grid()

labels = [(str(-12+i) + 'π/2' if i != 12 else '0') for i in range(25)]
plt.xticks(np.arange(-6*np.pi, 6*np.pi+np.pi/2, np.pi/2), labels)
ax.legend()
```

```python
#Часть 3
x = np.linspace(-np.pi,np.pi,150)
y = np.linspace(-np.pi,np.pi,150)
X,Y = np.meshgrid(x,y)

Z = np.sin(X) * np.sin(Y) 

fig, ax = plt.subplots(figsize=(6,5))

norm = mpl.colors.Normalize(-abs(Z).max(), abs(Z).max())

p = ax.pcolor(X, Y, Z, norm = norm, cmap=mpl.cm.cool)
ax.axis('tight')

```

```python
x = np.linspace(-3,3,150)
y = np.linspace(-3,3,150)
X,Y = np.meshgrid(x,y)

Z = np.exp(-X**2-Y**2) 

fig, ax = plt.subplots(figsize=(6,5))

norm = mpl.colors.Normalize(-abs(Z).max(), abs(Z).max())

p = ax.pcolor(X, Y, Z, norm = norm, cmap=mpl.cm.spring)
ax.axis('tight')
```
